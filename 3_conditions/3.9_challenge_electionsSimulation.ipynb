```python
from random import random


# set up a small function to return a probability
def generate_p():
    """returns a random float between 0.00 and 0.99"""
    return random()


def election_simulation():
    """simulates an election by calculating a winner based on probabilities"""

    # Three relevant regions
    r1 = .87
    r2 = .65
    r3 = .17

    # Two possible candidates
    a_wins = 0
    b_wins = 0

    # Run the simulation 10_000 times
    for trial in range(10_000):

        # Region counter for the simulation (initialized to 1)
        region_id = 1

        # loop over a region-range of 3
        while region_id <= 3:
            print(region_id)
            print(generate_p())

            # Evaluate if candidate A wins region 1
            if (region_id == 1) and (generate_p() <= r1):
                a_wins = a_wins + 1

            # Evaluate if candidate A wins region 2
            elif (region_id == 2) and (generate_p() <= r2):
                a_wins = a_wins + 1

            # Evaluate if candidate A wins region 3
            elif (region_id == 3) and (generate_p() <= r3):
                a_wins = a_wins + 1

            # If candidate A doesn't win, it must be candidate B
            else:
                b_wins = b_wins + 1

            print(f"win counter A: {a_wins}")
            print(f"win counter B: {b_wins}")
            region_id = region_id + 1

    if a_wins > b_wins:
        overall_winner = "The overall winner is: candidate A"
    else:
        overall_winner = "The overall winner is: candidate B"

    return overall_winner



# Run the function and print the contents
print(election_simulation())

```
