```python
# Do the necessary imports
import pandas as pd  # purpose: setting up a dataframe
import numpy as np  # purpose: use the linear regression method from that one
import matplotlib.pyplot as plt  # purpose: visualize the results

csv_data = pd.read_csv(r"C:\Users\Joshua Albert\Dropbox\PC\Downloads\cars_data.csv",  skiprows = 1, names = ['fuel_type', 'chassis_type', 'horse_power', 'age', 'price'])

x = list(csv_data['horse_power']) # age / alternatively hp
y = list(csv_data['price']) # price


# Create the dataframe using a dictionary (consisting of our x-axis and y-axis coordinates from above)
df = pd.DataFrame({'x': x, 'y': y})

# Actually call the method for calculating a linear regression
m = np.polyfit(df['x'], df['y'], 1)
#
f = np.poly1d(m)
#
df.insert(2, 'Fit', f(df['x']))

# Provide the required arguments for plotting the data points
plt.scatter(df['x'], df['y'], c="blue")

# Provide the required arguments for plotting the regression graph
plt.plot(df['x'], df['Fit'], c="orange")

# Define ranges for both axes
plt.xlim([min(x),max(x)])
plt.ylim([min(y), max(y)])
# Define labels for both axes
plt.xlabel('x')
plt.ylabel('y')

# Define a title for the resulting visual
plt.title('Linear Regression Example')
# Finally, show the plotted result
plt.show()
```
